```python
df.rename(columns={"availability_365": "days_booked"}).info()
```

```python
## Rename the column.
df.rename(columns={"cancellation_policy": "Policy_OfCancellation"}).info()
```

```python
## Convert all column names to lowercase and replace the spaces with an underscore "_"
df.columns = [x.lower() for x in df.columns]
df.columns = [x.replace(' ', '_') for x in df.columns]
df.head()
```

```python
##df = pd.to_numeric(df.replace('[^0-9\.-]', '', regex=True))
```

```python
## Remove the dollar sign and comma from the columns. If necessary, convert these two columns to the appropriate data type.
df["price"] = df["price"].str.replace("$", "",).str.replace(",", "",).astype(float)
```

```python
df["service_fee"] = df["service_fee"].str.replace("$", "").str.replace(",", "").head().astype(float)
```
